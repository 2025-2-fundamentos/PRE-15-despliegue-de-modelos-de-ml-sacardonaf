```python
import pickle

import pandas as pd  # type: ignore
from sklearn.linear_model import LinearRegression  # type: ignore

df = pd.read_csv("../files/input/house_data.csv", sep=",")

features = df[
    [
        "bedrooms",
        "bathrooms",
        "sqft_living",
        "sqft_lot",
        "floors",
        "waterfront",
        "condition",
    ]
]

target = df[["price"]]
```

```python
features.bathrooms.value_counts().sort_index().plot(kind="bar")
```

```python
features.sqft_living.plot(kind="hist", edgecolor="white", bins=20)
```

```python
features.sqft_lot.plot(kind="hist", edgecolor="white", bins=10)
```

```python
estimator = LinearRegression()
estimator.fit(features, target)

with open("../homework/house_predictor.pkl", "wb") as file:
    pickle.dump(estimator, file)
```
